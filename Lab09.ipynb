```python
import pandas as pd
import numpy as np

# Load the dataset
# Replace 'your_file_path.csv' with the actual path to your dataset
data = pd.read_csv('/content/household_power_consumption.txt.csv', sep=';', parse_dates=[['Date', 'Time']], na_values=[''])

# Rename the combined datetime column for convenience
data.rename(columns={'Date_Time': 'datetime'}, inplace=True)

# Display the first few rows of the dataset
print(data.head())

# Check for missing values
missing_values = data.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Handle missing values (e.g., forward fill)
data.fillna(method='ffill', inplace=True)

# Convert relevant columns to appropriate data types if necessary
# For example, if 'global_active_power' is in string format, convert it to float
data['global_active_power'] = data['global_active_power'].astype(float)

# Example of calculating active energy consumed by electrical equipment not measured
data['unmetered_power'] = (data['global_active_power'] * 1000 / 60) - data['sub_metering_1'] - data['sub_metering_2'] - data['sub_metering_3']

# Display summary statistics
print(data.describe())

# Save cleaned dataset to a new CSV file if needed
data.to_csv('cleaned_dataset.csv', index=False)
```
